```python
resumen = df.groupby("personaje").agg(
    partidas=("victoria", "count"),
    win_rate=("victoria", "mean"),
    promedio_minutos=("minutos", "mean"),
    promedio_oro=("oro", "mean"),
    resurrecciones=("resurrecciones_usadas", "mean")
).round(2)

resumen["win_rate"] = (resumen["win_rate"] * 100).round(1)
print(resumen)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("C:/Users/dcastan1/OneDrive - Clariant International Ltd/Desktop/VS_dataset.xlsx")
print(df.shape)
```

```python
resumen = df.groupby("personaje").agg(
    partidas=("victoria", "count"),
    win_rate=("victoria", "mean"),
    promedio_minutos=("minutos", "mean"),
    promedio_oro=("oro", "mean"),
    resurrecciones=("resurrecciones_usadas", "mean")
).round(2)

resumen["win_rate"] = (resumen["win_rate"] * 100).round(1)
print(resumen)
```

```python
resumen_mapa = df.groupby("mapa").agg(
    partidas=("victoria", "count"),
    win_rate=("victoria", "mean"),
    promedio_minutos=("minutos", "mean"),
    promedio_oro=("oro", "mean")
).round(2)

resumen_mapa["win_rate"] = (resumen_mapa["win_rate"] * 100).round(1)
print(resumen_mapa)
```

```python
resumen_combo = df.groupby(["mapa", "personaje"]).agg(
    partidas=("victoria", "count"),
    win_rate=("victoria", "mean"),
    promedio_oro=("oro", "mean")
).round(2)

resumen_combo["win_rate"] = (resumen_combo["win_rate"] * 100).round(1)
print(resumen_combo)
```
