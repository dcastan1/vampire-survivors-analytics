```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("C:/Users/dcastan1/OneDrive - Clariant International Ltd/Desktop/VS_dataset.xlsx")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Mi Dashboard - Vampire Survivors", fontsize=16, fontweight="bold")

# Gráfica 1 - Win rate por personaje
win_personaje = df.groupby("personaje")["victoria"].mean() * 100

axes[0, 0].bar(win_personaje.index, win_personaje.values, color="purple")
axes[0, 0].set_title("Win Rate por Personaje")
axes[0, 0].set_xlabel("Personaje")
axes[0, 0].set_ylabel("Win Rate %")
axes[0, 0].tick_params(axis="x", rotation=45)

# Gráfica 2 - Promedio de oro por mapa
oro_mapa = df.groupby("mapa")["oro"].mean()

axes[0, 1].bar(oro_mapa.index, oro_mapa.values, color="gold")
axes[0, 1].set_title("Promedio de Oro por Mapa")
axes[0, 1].set_xlabel("Mapa")
axes[0, 1].set_ylabel("Oro promedio")
axes[0, 1].tick_params(axis="x", rotation=45)

# Gráfica 3 - Minutos promedio por personaje
minutos_personaje = df.groupby("personaje")["minutos"].mean()

axes[1, 0].bar(minutos_personaje.index, minutos_personaje.values, color="steelblue")
axes[1, 0].set_title("Promedio de Minutos por Personaje")
axes[1, 0].set_xlabel("Personaje")
axes[1, 0].set_ylabel("Minutos promedio")
axes[1, 0].tick_params(axis="x", rotation=45)

# Gráfica 4 - Resurrecciones promedio por personaje
resurrecciones = df.groupby("personaje")["resurrecciones_usadas"].mean()

axes[1, 1].bar(resurrecciones.index, resurrecciones.values, color="crimson")
axes[1, 1].set_title("Resurrecciones Promedio por Personaje")
axes[1, 1].set_xlabel("Personaje")
axes[1, 1].set_ylabel("Resurrecciones promedio")
axes[1, 1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()
```
