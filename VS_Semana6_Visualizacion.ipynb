```python
import pandas as pd
import matplotlib.pyplot as plt

partidas = [
    {"jugador": "Carlos", "personaje": "Mortaccio", "minutos": 18, "enemigos": 4823, "victoria": True},
    {"jugador": "María", "personaje": "Imelda", "minutos": 12, "enemigos": 2100, "victoria": False},
    {"jugador": "Luis", "personaje": "Clerici", "minutos": 25, "enemigos": 7654, "victoria": True},
    {"jugador": "Ana", "personaje": "Mortaccio", "minutos": 8, "enemigos": 1200, "victoria": False},
    {"jugador": "Carlos", "personaje": "Clerici", "minutos": 30, "enemigos": 9100, "victoria": True}
]

df = pd.DataFrame(partidas)
```

```python
df.plot(kind="bar", x="jugador", y="enemigos", color="purple")
plt.title("Enemigos eliminados por jugador")
plt.xlabel("Jugador")
plt.ylabel("Enemigos")
plt.show()
```

```python
df_agrupado = df.groupby("jugador")["enemigos"].sum()
print(df_agrupado)
```

```python
df_agrupado.plot(kind="bar", color="purple")
plt.title("Total de enemigos eliminados por jugador")
plt.xlabel("Jugador")
plt.ylabel("Enemigos totales")
plt.xticks(rotation=0)
plt.show()
```

```python
df_agrupado2 = df.groupby("jugador")["enemigos"].mean()
print(df_agrupado2)
```

```python
df_agrupado.plot(kind="bar", color="purple")
plt.title("Total de enemigos eliminados por jugador")
plt.xlabel("Jugador")
plt.ylabel(df_agrupado2)
plt.xticks(rotation=0)
plt.show()
```

```python
df_agrupado2 = df.groupby("personaje")["enemigos"].mean()

df_agrupado2.plot(kind="bar", color="purple")
plt.title("Promedio de enemigos por personaje")
plt.xlabel("Personaje")
plt.ylabel("Promedio de enemigos")
plt.xticks(rotation=0)
plt.show()
```
