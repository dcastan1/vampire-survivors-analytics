```python
import pandas as pd
import matplotlib.pyplot as plt

partidas = [
    {"jugador": "Carlos", "personaje": "Arcano", "minutos": 22, "enemigos": 6100, "oro": 450, "victoria": True},
    {"jugador": "María", "personaje": "Sombra", "minutos": 14, "enemigos": 3200, "oro": 280, "victoria": False},
    {"jugador": "Luis", "personaje": "Arcano", "minutos": 28, "enemigos": 8400, "oro": 620, "victoria": True},
    {"jugador": "Ana", "personaje": "Titan", "minutos": 19, "enemigos": 4900, "oro": 390, "victoria": True},
    {"jugador": "Carlos", "personaje": "Sombra", "minutos": 11, "enemigos": 2800, "oro": 210, "victoria": False},
    {"jugador": "María", "personaje": "Titan", "minutos": 24, "enemigos": 5600, "oro": 480, "victoria": True},
    {"jugador": "Luis", "personaje": "Espectro", "minutos": 16, "enemigos": 3900, "oro": 320, "victoria": False},
    {"jugador": "Ana", "personaje": "Espectro", "minutos": 20, "enemigos": 5100, "oro": 410, "victoria": True},
    {"jugador": "Carlos", "personaje": "Titan", "minutos": 31, "enemigos": 9200, "oro": 710, "victoria": True},
    {"jugador": "María", "personaje": "Arcano", "minutos": 18, "enemigos": 4700, "oro": 380, "victoria": False}
]

df = pd.DataFrame(partidas)
```

```python
print(df)
```

```python
#Agrupamiento por personaje
resumen=df.groupby("personaje").agg(
    partidas_jugadas=("victoria","count"),
    victorias=("victoria","sum"),
    promedio_minutos=("minutos","mean"),
    promedio_oro=("oro","mean")).round(2)

resumen["Win_rate"]=(resumen["victorias"]/resumen["partidas_jugadas"]*100).round(2)

print(resumen)
```

```python
resumen[["promedio_oro","promedio_minutos"]].plot(kind="bar")
plt.title("Comparación de personajes - Minutos vs Oro")
plt.xlabel("personaje")
plt.xticks(rotation=0)
plt.show()
```
