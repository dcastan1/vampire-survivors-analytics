```python
partidas = [
    {"jugador": "Carlos", "personaje": "Mortaccio", "minutos": 18, "enemigos": 4823, "victoria": True},
    {"jugador": "María", "personaje": "Imelda", "minutos": 12, "enemigos": 2100, "victoria": False},
    {"jugador": "Luis", "personaje": "Clerici", "minutos": 25, "enemigos": 7654, "victoria": True},
    {"jugador": "Ana", "personaje": "Mortaccio", "minutos": 8, "enemigos": 1200, "victoria": False},
    {"jugador": "Carlos", "personaje": "Clerici", "minutos": 30, "enemigos": 9100, "victoria": True}
]
```

```python
for partida in partidas:
    if partida["victoria"]==True:
        print(partida["jugador"], "jugó con", partida["personaje"], "durante", partida ["minutos"], "minutos", "- VICTORIA")
    else:
         print(partida["jugador"], "jugó con", partida["personaje"], "durante", partida ["minutos"], "minutos", "- DERROTA")
```

```python
victoria=0
derrotas=0
for partida in partidas:
    if partida ["victoria"] == True: 
     victoria = victoria + 1
    else:
     derrotas = derrotas + 1

print ("victorias:", victoria)
print ("derrotas:", derrotas)
print ("Win Rate", victoria/(victoria+derrotas)*100, "%")
```

```python
maximo=0

for partida in partidas:
    if partida["enemigos"]>maximo:
        maximo = partida ["enemigos"]
for partida in partidas:
    if maximo == partida ["enemigos"]:
        print("El jugador es:", partida["jugador"], "con", maximo, "enemigos eliminados")
```
