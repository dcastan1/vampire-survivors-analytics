```python
partidas = [
    {"personaje": "Mortaccio", "minutos": 18, "enemigos": 4823, "victoria": True},
    {"personaje": "Imelda", "minutos": 12, "enemigos": 2100, "victoria": False},
    {"personaje": "Clerici", "minutos": 25, "enemigos": 7654, "victoria": True}
]
```

```python
for partida in partidas:
    print(partida["personaje"])
```

```python
for partida in partidas:
    print(partida["personaje"], "-", partida["minutos"], "minutos -", partida["enemigos"], "enemigos")
```

```python
for partida in partidas:
    if partida["victoria"] == True:
        print(partida["personaje"], "- GANÓ")
    else:
        print(partida["personaje"], "- perdió")
```

```python
victorias = 0
derrotas = 0

for partida in partidas:
    if partida["victoria"] == True:
        victorias = victorias + 1
    else:
        derrotas = derrotas + 1

print("Victorias:", victorias)
print("Derrotas:", derrotas)
print("Win rate:", victorias / (victorias + derrotas) * 100, "%")
```
