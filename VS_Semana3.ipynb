```python
partida = {
    "personaje": "Mortaccio",
    "minutos": 18,
    "enemigos": 4823,
    "victoria": True
}

print(partida)
print(partida["personaje"])
print(partida["minutos"])
```

```python
partida["armas"] = ["Whip", "Garlic", "Holy Water"]

print(partida)
print(partida["armas"])
print(partida["armas"][0])
```

```python
def resumen_partida(p):
    print("Personaje:", p["personaje"])
    print("Duración:", p["minutos"], "minutos")
    print("Enemigos eliminados:", p["enemigos"])
    print("Victoria:", p["victoria"])
    print("Armas usadas:", len(p["armas"]))

resumen_partida(partida)
```

```python
partida2 = {
    "personaje": "Imelda",
    "minutos": 12,
    "enemigos": 2100,
    "victoria": False,
    "armas": ["Fire Wand", "Santa Water"]
}

resumen_partida(partida)
resumen_partida(partida2)
```
